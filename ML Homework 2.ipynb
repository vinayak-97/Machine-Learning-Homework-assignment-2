```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(df[["Age","Fare"]])

df[["Age","Fare"]] = scaler.transform(df[["Age","Fare"]])

df
```

```python
X = df.drop("Survived",axis=1)
y = df["Survived"]
```

```python
from sklearn.tree import DecisionTreeClassifier

Dt_model = DecisionTreeClassifier()
Dt_model.fit(X,y)
```

```python
dt_score = round(Dt_model.score(X,y)*100,2)
print("Decision Tree score: ",dt_score)
```

```python
df_test = pd.read_csv("test.csv")
df_test.head(2)
```

```python
df_test.drop(["PassengerId","Name","Ticket","Cabin"],axis=1,inplace=True)
df_test.head(2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score


dt = DecisionTreeClassifier(random_state=42)

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, 9, 11],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5)


grid_search.fit(X, y)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print('Best hyperparameters:', best_params)
print('Best classification accuracy:', best_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

best_dt_model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 9, 
                                       min_samples_leaf = 2, 
                                       min_samples_split = 10)
best_dt_model.fit(X,y)
best_dt_model.score(X,y)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 10))
plot_tree(best_dt_model, filled=True, ax=ax, 
          feature_names=X.columns, 
          class_names=['Not Survived', 'Survived'],
          max_depth=4,fontsize=10,rotate=True)
plt.show()
```

```python
#Apply the five-fold cross validation of your fine-tuned decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier(criterion ='entropy', max_depth = 9, 
                             min_samples_leaf = 2, min_samples_split = 10)

scores = cross_val_score(clf, X, y, cv=5)
print(scores)
print('Average classification accuracy:', scores.mean())
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

Rf_model = RandomForestClassifier()
Rf_model.fit(X,y)
Rf_score = round(Rf_model.score(X,y)*100,2)
print("Random Forest Score: ",Rf_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)

grid_search.fit(X, y)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print('Best hyperparameters:', best_params)
print('Best classification accuracy:', best_score)
```

```python
rf_model = RandomForestClassifier(max_depth = None, min_samples_leaf = 2, 
                            min_samples_split = 10, n_estimators = 100)

rf_model.fit(X,y)
rf_model.score(X,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rf = RandomForestClassifier(max_depth = None, min_samples_leaf = 2, 
                            min_samples_split = 10, n_estimators = 100)

scores = cross_val_score(rf, X, y, cv=5)
print(scores)
avg_score = scores.mean()
print("Random Forest Average Classification Accuracy:", avg_score)
```
